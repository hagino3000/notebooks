```python
import scipy.stats
```

```python
x = np.arange(0, 21)
y = scipy.stats.binom.pmf(x, 20, 0.5)
```

```python
plt.figure(figsize=(8, 2))
plt.bar(x, y)
plt.xlabel('表が出る回数')
plt.ylabel('確率')
```

```python
import pandas as pd
p_value = pd.DataFrame({'表の出る回数':x, '確率': y}).query(
    '表の出る回数 >= 15'
)['確率'].sum()
print('{:,.4f}'.format(p_value))
```

```python
x = np.linspace(0, 1, 200)

# 流入元がA
n = 205
converted = 35
p = converted/n
std = np.sqrt(p*(1-p)/n)
y_a = scipy.stats.norm.pdf(x, p, std)

# 流入元がB
n = 290
converted = 70
p = converted/n
std = np.sqrt(p*(1-p)/n)
y_b = scipy.stats.norm.pdf(x, p, std)
```

```python
plt.figure(figsize=(7, 2))
plt.plot(x, y_a, label='A')
plt.plot(x, y_b, label='B')
plt.legend(loc='best')
plt.xlabel('流入ユーザーの継続化率')
plt.ylabel('確率')
```

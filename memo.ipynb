X, Y の同時分布の確率密度関数が

fX,Y (x, y) =  {
e.exp(-x-y) ( x >= 0 and y >= 0)
0                 (他)
}
だったとする。以下の確率を求めよ。
1. 0 <= X <= 1 and 0 <= Y <= 1
2. X >= 1 and Y >= 1

```python
from sympy.abc import x,y
fxy = sym.exp(-x-y)
print(fxy)
sym.integrate(fxy, x)
```

```python
fig, ax = plt.subplots()
import numpy as np
x = np.linspace(-1 * np.pi, np.pi, 100)

ax.plot(x, np.sin(x), 'rd', label="Sin(x)")
ax.plot(x, np.cos(x), label="Cos(x)")
plt.xlabel('X')
plt.ylabel('Y')
ax.legend()
```

```python
X = np.linspace(-1*np.pi, np.pi, 100)
Y = np.linspace(-1*np.pi, np.pi, 100)

Z = np.zeros([100, 100])
for xi, x in enumerate(X):
    for yi, y in enumerate(Y):
        Z[xi, yi] = np.cos(x) + np.cos(y)
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
fig = plt.figure(figsize=(12,6))

# `ax` is a 3D-aware axis instance because of the projection='3d' keyword argument to add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# surface_plot with color grading and color bar
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1, 1, 1, projection='3d')
p = ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4)
```

```python
lat = np.linspace(-0.5 * np.pi, 0.5 * np.pi, 1000)
```

```python
earth_r = 6371
```

```python
lat_r = lat/np.pi*180
```

```python
plt.plot(lat_r, np.sin(lat))
```

```python
plt.plot(lat_r, np.arctanh(np.sin(lat)) * earth_r)
plt.xlabel('Latitude')
plt.ylabel('Distance (km)')
plt.title(u'')
```
